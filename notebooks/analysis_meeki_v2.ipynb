```python
import pandas as pd
import os

# Path to the specific file you found on GitHub
file_path = os.path.join("..", "data", "processed", "unique_clean_df.csv")

# Load the data
df = pd.read_csv(file_path)

# Check the first few rows and the column count
print(f"Columns found: {len(df.columns)}")
df.head()
print(len(df))
```

**Data Importing and Pre-processing**

**Import dataset and describe characteristics such as dimensions, data types, file types, and import methods used**
**Transform data appropriately using techniques such as aggregation, normalization, and feature construction**


```python
# Create House Age
df['date'] = pd.to_datetime(df['date'])
df['house_age'] = df['date'].dt.year - df['yr_built']

# Normalize Sqft
df['sqft_living_norm'] = (df['sqft_living'] - df['sqft_living'].min()) / \
                         (df['sqft_living'].max() - df['sqft_living'].min())

print("Construction and Normalization complete.")
df.head()
print(len(df))
df.shape
```

```python
# AGGREGATION: Group by house grade to see the average price
# This satisfies the "Aggregation" requirement
agg_df = df.groupby('grade')['price'].mean().reset_index()
print("Aggregation Complete: Average Price per Grade")
agg_df
```

```python
# Need-Based Discretization (Price Bins)
df['price_bin'] = pd.cut(df['price'], bins=3, labels=['Budget', 'Mid-Range', 'Luxury'])

# Reduce Redundancy (Drop yr_built now that we have house_age)
df = df.drop(columns=['yr_built'])

# The Final Result
print("FINAL SHAPE:", df.shape) 
# Target: (21436, 23)
```

```python

df.to_csv('../data/processed/final_wrangled_data_meeki.csv', index=False)

print("Success! File saved to the processed data folder.")
```
