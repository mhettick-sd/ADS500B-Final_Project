As part of our pre-processing, we will need to convert this to a proper datetime format to allow for time-based analysis.


```python
print("--- Missing Values Per Column ---")
print(houses.isnull().sum())
```

**Clean, wrangle, and handle missing data**

```python
# 1. Handle Missing Data: Fill with Median

target_columns = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot']
for col in target_columns:
    houses[col] = houses[col].fillna(houses[col].median())
```

```python
# 2. Wrangle Data: Convert date string to datetime object
houses['date'] = pd.to_datetime(houses['date'])
```

```python
df = houses
```

```python
print(df.isnull().sum())
```

**Handling Missing Data**: Our initial scan revealed 1,134 missing values in bedrooms and similar gaps in bathrooms and sqft_living. By applying median imputation, we filled these gaps using a central tendency measure that is resistant to outliers. This ensures that extreme "mansion" prices in the King County dataset do not unfairly influence the replacement values for missing data, maintaining a more realistic distribution for the majority of homes.

**Data Wrangling (Type Conversion)**: Converting the date column from a string to a datetime object is the most significant wrangling step. In its original format, the date was non-calculable text. As a datetime object, we can now extract specific time features (year, month, day) and calculate the Age of the House at Sale, which is a high-impact feature for price prediction.

The post-cleaning verification confirms a 0% null rate across all 21,613 rows. The dataset is now structurally ready for the exploratory data analysis (EDA) phase.
